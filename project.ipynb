```python
df=df.drop_duplicates()# if duplicates are present then use this
```

```python
# Removing outliers
import numpy as np
```

```python
numeric_col= df.select_dtypes(include=[np.number])
```

```python
numeric_col
```

```python
Q1=numeric_col.quantile(0.25)
Q3=numeric_col.quantile(0.75)
IQR= Q3-Q1
```

```python
lower= Q1-1.5*IQR
upper= Q3+1.5*IQR
```

```python
Outliers= (numeric_col<lower)|(numeric_col>upper)
```

```python
Outliers_per_col=Outliers.sum()
Outliers_per_rows=Outliers.any(axis=1).sum()
```

```python
df_filtered = df[~Outliers.any(axis=1)]
```
